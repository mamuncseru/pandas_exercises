## Filtering DataFrames by many conditions (AND)

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
mask1 = titanic.sex == 'male'
mask1.head()
```

```python
mask2 = titanic.age > 14
mask2.head()
```

```python
(mask1 & mask2).head()
```

```python
male_surv = titanic.loc[mask1 & mask2, ["survived", "pclass", "sex", "age"]]
male_surv.head(20)
```

```python
male_surv.info()
```

```python
male_surv.describe()
```

```python
titanic.describe()
```

## Filtering DataFrames with many Conditions (OR)

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head(10)
```

```python
mask1 = titanic.sex == "female"
mask1.head(20)
```

```python
mask2 = titanic.age < 14
mask2.head(20)
```

```python
(mask1 | mask2).head(20)
```

```python
titanic.loc[mask1 | mask2]
```

```python
wom_or_chi = titanic.loc[mask1 | mask2, ["survived", "pclass", "sex", "age"]]
```

```python
wom_or_chi.info()
```

```python
wom_or_chi.describe()
```

```python
titanic.describe()
```

## Advanced Filtering with between(), isin() and ~

```python
import pandas as pd
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head()
```

```python
og_1988 = summer.loc[summer.Year == 1988]
```

```python
og_1988.head()
```

```python
og_1988.tail()
```

```python
og_1988.info()
```

```python
og_since1992 = summer.loc[summer.Year >= 1992]
```

```python
og_since1992.head()
```

```python
og_since1992.tail()
```

```python
summer.Year.between(1960, 1969).head()
```

```python
og_60s = summer.loc[summer.Year.between(1960, 1969, inclusive=True)]
```

```python
og_60s.head()
```

```python
og_60s.tail()
```

```python
my_favourite_games = [1972, 1996]
```

```python
summer.Year.isin(my_favourite_games).head()
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_favourite_games)]
```

```python
og_72_96.head()
```

```python
og_72_96.tail()
```

```python
og_not_72_96 = summer.loc[~summer.Year.isin(my_favourite_games)]
```

```python
og_not_72_96.head()
```

```python
og_not_72_96.Year.unique()
```
