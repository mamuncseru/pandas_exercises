## Changing a single value (Option 1 with loc)

```python
titanic.loc[1, 'age'] = 40
```

```python
titanic.head()
```

```python
titanic.iloc[1, 3] = 41 # Changine a single value with iloc
```

```python
titanic.head()
```

### Changing multiple values in a column with loc

```python
titanic.loc[1:3, "age"] = 42
```

```python
titanic.head()
```

### Changing multiple values in a column with iloc

```python
titanic.iloc[1:4, 3] = [43, 44, 45]
```

```python
titanic.head()
```

### Changing multiple values in a column with boolean indexing

```python
index_babies = titanic.loc[titanic.age < 1, "age"].index
```

```python
titanic.loc[titanic.age < 1, "age"] = 1
```

```python
titanic.loc[index_babies]
```

### Changing multiple values in a row

```python
titanic.head()
```

```python
titanic.head()
```

### Changing multiple values in multiple rows/columns

```python
titanic.replace(0, "Zero")
```

## View Vs. Copy

### Slicing a DataFrame / creating a view on the original DataFrame

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
age = titanic.age
```

```python
age.head()
```

```python
age._is_view
```

```python
age._is_copy is None
```

```python
age[1] = 40
```

```python
age.head()
```

```python
titanic.head()
```

## Slicing a DataFrame / creating a copy of the original DataFrame

```python
df_baby = titanic[titanic.age < 1]
```

```python
df_baby._is_view
```

```python
df_baby._is_copy is None
```

```python
df_baby._is_copy()
```

```python
df_baby.age = 1
```

```python
df_baby
```

```python
titanic.loc[index_babies]
```

### If you want to work and manipulate the Slice DataFrame...
### ... and make a copy with .copy()!!!

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
age = titanic.age.copy()
```

```python
age[1] = 40
```

```python
age.head()
```

```python
titanic.head()
```

```python
baby_ages = titanic.loc[titanic.age < 1, ["age", "sex"]].copy()
```

```python
baby_ages
```

```python
baby_ages["age"] = 1
```

```python
baby_ages
```

```python
titanic.loc[index_babies]
```
